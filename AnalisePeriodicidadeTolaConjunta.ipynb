```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

subpath_raw = "naive_periodic_experiment/2025-01-06T19:07:27"
data_raw = pd.read_csv(f'{subpath_raw}/total_X.csv')
data_raw['Experimento'] = 'Periodicidade Tola'

subpath_fs = "naive_periodic_experiment/2025-01-06T19:07:27"
data_fs = pd.read_csv(f'{subpath_fs}/minimal_with_univariate.csv')
data_fs['Experimento'] = 'Periodicidade Tola FS'

subpath_tptt = "naive_periodic_experiment_tptt/2025-01-26T20:09:01"
data_tptt = pd.read_csv(f'{subpath_tptt}/total_X.csv')
data_tptt['Experimento'] = 'Periodicidade Tola TPTT'

subpath_fs_tptt = "naive_periodic_experiment_tptt/2025-01-26T20:09:01"
data_fs_tptt = pd.read_csv(f'{subpath_fs_tptt}/minimal_with_univariate.csv', index_col=False)
data_fs_tptt['Experimento'] = 'Periodicidade Tola FS TPTT'

data_combined = pd.concat([data_raw, data_fs, data_tptt, data_fs_tptt], ignore_index=True)

data_combined.drop(columns=['Unnamed: 6'], inplace=True, errors='ignore')
y_limits = (0, 0.25)

plt.figure(figsize=(14, 8))
sns.boxplot(x='perodo', y='nmae', hue='Experimento', data=data_combined, palette="muted")
plt.rcParams['font.size'] = 16

plt.xlabel("Perodo")
plt.ylabel("NMAE")
#plt.title("Comparao do NMAE por Perodo e Experimento", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.ylim(y_limits)
plt.show()
```
