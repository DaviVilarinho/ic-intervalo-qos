```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

subpath_distrib = "distrib_experiment_y_original/2025-01-05T17:06:02"
data_distrib = pd.read_csv(f'{subpath_distrib}/total_X.csv')
data_distrib['Experimento'] = 'Janelas de Distribuições'

subpath_distrib_fs = "distrib_experiment_y_original/2025-01-05T17:06:02"
data_distrib_fs = pd.read_csv(f'{subpath_distrib_fs}/minimal_with_univariate.csv')
data_distrib_fs['Experimento'] = 'Janelas de Distribuições FS'

data_combined = pd.concat([data_distrib, data_distrib_fs], ignore_index=True)

data_combined.drop(columns=['Unnamed: 7'], inplace=True, errors='ignore')

plt.figure(figsize=(14, 8))
sns.boxplot(x='período', y='nmae', hue='Experimento', data=data_combined, palette="muted")

plt.xlabel("Período")
plt.ylabel("NMAE")
plt.xticks(rotation=45)
plt.tight_layout()
y_limits = (0, 0.25)
plt.ylim(y_limits)

plt.rcParams['font.size'] = 16

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

subpath_distrib = "distrib_experiment_y_original/2025-01-05T17:06:02"
data_distrib = pd.read_csv(f'{subpath_distrib}/total_X.csv')
data_distrib['Experimento'] = 'Janelas de Distribuições'

subpath_distrib_fs = "distrib_experiment_y_original/2025-01-05T17:06:02"
data_distrib_fs = pd.read_csv(f'{subpath_distrib_fs}/minimal_with_univariate.csv')
data_distrib_fs['Experimento'] = 'Janelas de Distribuições FS'

data_combined = pd.concat([data_distrib, data_distrib_fs], ignore_index=True)

data_combined.drop(columns=['Unnamed: 7'], inplace=True, errors='ignore')

data_combined['Método e FS'] = data_combined['método'] + ' - ' + data_combined['Experimento']

plt.figure(figsize=(14, 8))
sns.boxplot(x='período', y='nmae', hue='Método e FS', data=data_combined, palette="muted")

plt.xlabel("Período")
plt.ylabel("NMAE")
plt.xticks(rotation=45)
plt.tight_layout()
y_limits = (0, 0.25)
plt.ylim(y_limits)

plt.rcParams['font.size'] = 16

plt.show()
```
