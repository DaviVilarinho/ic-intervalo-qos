```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

j = 16
tempo_total = 256

tempo_y = np.arange(0, tempo_total, j)
qos_y = np.random.randint(13, 25, len(tempo_y))

tempo_tptt = np.arange(0, tempo_total)
qos_tptt = np.random.randint(13, 25, tempo_total)

df_y = pd.DataFrame({'Tempo': tempo_y, 'DispFrames': qos_y, 'Tipo': 'Y\': utilizado em treinamento'})
df_tptt = pd.DataFrame({'Tempo': tempo_tptt, 'DispFrames': qos_tptt, 'Tipo': 'TPTT: utilizado em testes'})

df = pd.concat([df_y, df_tptt])

plt.figure(figsize=(10, 6))
for tipo, dados in df.groupby('Tipo'):
    plt.plot(dados['Tempo'], dados['DispFrames'], marker='o', label=tipo, linestyle=' ')

plt.xlim(16,256)
plt.title('Comparação entre Y\' e TPTT')
plt.xlabel('Tempo (s)')
plt.ylabel('Quadros por Segundo')
plt.legend()
plt.grid(True)
plt.show()
```
