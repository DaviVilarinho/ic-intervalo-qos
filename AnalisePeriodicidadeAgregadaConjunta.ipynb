```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

subpath_agg = "agg_function_periodic_experiment_y_original/2025-01-06T17:07:48"
data_agg = pd.read_csv(f'{subpath_agg}/total_X.csv')
data_agg['Experimento'] = 'Janelas de Funções'

subpath_agg_fs = "agg_function_periodic_experiment_y_original/2025-01-06T17:07:48"
data_agg_fs = pd.read_csv(f'{subpath_agg_fs}/minimal_with_univariate.csv')
data_agg_fs['Experimento'] = 'Janelas de Funções FS'

data_combined = pd.concat([data_agg, data_agg_fs], ignore_index=True)

data_combined.drop(columns=['Unnamed: 7'], inplace=True, errors='ignore')

plt.figure(figsize=(14, 8))
sns.boxplot(x='período', y='nmae', hue='Experimento', data=data_combined, palette="muted")

plt.xlabel("Período")
plt.ylabel("NMAE")
plt.xticks(rotation=45)
plt.tight_layout()
y_limits = (0, 0.25)
plt.ylim(y_limits)

plt.rcParams['font.size'] = 16

plt.show()
```
